```python
def quant(column, data=data, figsize=(5,5), error=False, ylim=False, show=True):
    if error==False:
        q1 = np.percentile(data[column], 25)
        q3 = np.percentile(data[column], 75)
    else:
        q3 = data[column].describe()['75%'] 
        q1 = data[column].describe()['25%']
    iqr = q3 - q1
    new_data = data[(data[column]>(q1-1.5*iqr))&(data[column]<(q3+1.5*iqr))] 
    if ylim!=False:
        new_dataplot.set_ylim(ylim)
    if show == True:
        new_dataplot = new_data.boxplot(column=column, figsize=figsize)
        display()
    return new_data
```

```python
def check(model, features, target):
    #PR
    probabilities = model.predict_proba(features)[:, 1]
    precision, recall, thresholds = precision_recall_curve(target, probabilities)
    plt.figure(figsize=(6, 6))
    plt.step(recall, precision, where='post')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('PR-кривая')
    plt.show()
    #ROC
    fpr, tpr, thresholds = roc_curve(target, probabilities) 
    plt.plot(fpr, tpr)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    #test-ROC
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0]) 
    plt.ylim([0.0, 1.0])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая")
    plt.show();
    auc_roc = roc_auc_score(target, probabilities)
    print('AUC-ROC =', auc_roc)
    #ConfusionMatrix
    predictions = model.predict(features)
    ConfusionMatrixDisplay(confusion_matrix(target, predictions)).plot();
    plt.title("ConfusionMatrix")
    plt.show();
    #ClassificationReport
    print(classification_report(target, predictions))
```

ROC-кривая - это кривая ошибок, чем график выше, тем выше верных значений.
AUC-ROC - это площадь ROC-кривой. AUC-ROC случайной модели равна 0.5, AUC-ROC нашей модели должна быть больше.

recall - процент правильных положительных прогнозов по отношению к общему количеству фактических положительных результатов.
f1-score - средневзвешенное гармоническое значение точности и полноты. Чем ближе к 1, тем лучше модель.

```python
%%time
wood_prediction_downsampled = None
wood_best_model_downsampled = None
wood_best_result_downsampled = 0
wood_best_est_downsampled = 0
wood_best_depth_downsampled = 0

for est in range(1, 51):
    model_plot_v = []
    model_plot_i = []
    for depth in range (1, 31):
        model = RandomForestClassifier(random_state=RANDOM_STATE, n_estimators=est, max_depth=depth, class_weight='balanced') 
        model.fit(features_downsampled, target_downsampled) 
        predictions_valid = model.predict(features_valid)
        result = f1_score(target_valid, predictions_valid)
        if result > wood_best_result_downsampled:
            wood_prediction_downsampled = predictions_valid
            wood_best_model_downsampled = model
            wood_best_result_downsampled = result
            wood_best_est_downsampled = est
            wood_best_depth_downsampled = depth
        model_plot_v.append(result)
        model_plot_i.append(depth)
    pd.Series(model_plot_v, index=model_plot_i).plot(grid=True)
print("F1 наилучшей модели на валидационной выборке:", wood_best_result_downsampled, "Количество деревьев:", wood_best_est_downsampled, "Максимальная глубина:", wood_best_depth_downsampled)
plt.title('Успешность модели случайного леса в зависимости от глубины')
plt.xlabel("Глубина леса")
plt.ylabel("Accuracy")
plt.show()
```
